```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/cumulative_2025.09.21_21.46.46.csv',skiprows=53)
#print(df)
s=[3,4]
y=df.iloc[:,s]
x=df.drop(df.columns[s],axis=1)
#print(x)

#print(len(df[df['koi_pdisposition']=='CONFIRMED']))
#print("Classes:", df['koi_tce_delivname'].unique())

x=(df['koi_disposition']!='CONFIRMED') & (df['koi_pdisposition']=='CANDIDATE')
y=(df['koi_disposition']=='CANDIDATE') & (df['koi_pdisposition']=='FALSEPOSITIVE')
z=(df['koi_disposition']=='CONFIRMED')&(df['koi_pdisposition']=='FALSEPOSITIVE')
#print(df[z])
df['selection']=np.select([(df['koi_disposition']=='CONFIRMED')& (df['koi_pdisposition']=='CANDIDATE'),
(df['koi_disposition']=='CANDIDATE')& (df['koi_pdisposition']=='CANDIDATE')],[1,2],0)
target=df['selection']
features=df.drop(columns=['selection','koi_disposition','koi_pdisposition','kepid','kepoi_name','kepler_name','koi_score','koi_teq_err1','koi_teq_err2','koi_fpflag_nt','koi_fpflag_ss','koi_fpflag_co','koi_fpflag_ec'])
#print(features)
#features = features.drop(columns=['koi_score'])
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
#cols=features['koi_tce_delivname']
hii=ColumnTransformer(transformers=[('hello',OneHotEncoder(),['koi_tce_delivname'])],remainder='passthrough')
encoded = hii.fit_transform(features)
features = pd.DataFrame(encoded, columns=hii.get_feature_names_out(), index=features.index)
#df['koi_tce_delivname'].head()
from sklearn.impute import SimpleImputer
numeric_cols = features.select_dtypes(include=np.number).columns
imputer=SimpleImputer(strategy='mean')
imputed_array = imputer.fit_transform(features[numeric_cols])
imputed_df = pd.DataFrame(imputed_array, columns=numeric_cols, index=features.index)
# Replace only the numeric columns
features[numeric_cols] = imputed_df
#features.info()
from sklearn.model_selection import train_test_split
features_train,features_test,target_train,target_test=train_test_split(features,target,stratify=target,test_size=0.2,random_state=200)
import lightgbm as lgb
from sklearn.metrics import accuracy_score
conditions={
    'objective':'multiclass',
    'metric':'multi_logloss',
    'num_class':3,
    'boosting_type':'gbdt',
    'num_leaves':100,
    'learning_rate':0.05,
   'feature_fraction':0.7,
    'bagging_fraction':0.6,
    'bagging_freq':1,
    'verbose':-1
}

model=lgb.LGBMClassifier(**conditions)
#Now, we have to onle handle categorical data and missing values,then our work will be done.
model.fit(features_train,target_train)
target_pred=model.predict(features_test)
print(accuracy_score(target_test,target_pred))
#print(df['selection'])
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(target_test, target_pred))
sns.heatmap(confusion_matrix(target_test, target_pred), annot=True, fmt='d')
plt.title("Confusion Matrix")
plt.show()

corr = pd.concat([features, target], axis=1).corr()
sns.heatmap(corr[['selection']].drop('selection'), annot=True)
plt.title("Feature Correlation with Target")
plt.show()
#koi_score=float(input("Disposition score : "))
#koi_fpflag_nt=float(input("not transit fpflag : "))
#koi_fpflag_ss=float(input("stellar eclipse fpflag : "))
#koi_fpflag_co=float(input("centroid offset fpflag"))
#koi_fpflag_ec=float(input("contamination fpflag"))
koi_period=float(input("Orbital Period(days) : "))
koi_period_err1=float(input("Orbital Period upper unc. : "))
koi_period_err2=float(input("Orbital Period lower unc. : "))
koi_time0bk=float(input("transit Epoch : "))
koi_time0bk_err1=float(input("transit Epoch upper unc. : "))
koi_time0bk_err2=float(input("transit Epoch lower unc. : "))
koi_impact=float(input("Impact Parameter : "))
koi_impact_err1=float(input("Impact Parameter upper unc. : "))
koi_impact_err2=float(input("Impact Parameter lower unc. : "))
koi_duration=float(input("Transit Duration(hours) : "))
koi_duration_err1=float(input("Transit Duration upper unc.(hours) : "))
koi_duration_err2=float(input("Transit Duration lower unc.(hours) : "))
koi_depth=float(input("Transit Depth(ppm) : "))
koi_depth_err1=float(input("Transit Depth upper unc.(ppm) : "))
koi_depth_err2=float(input("Transit Depth lower unc.(ppm) : "))
koi_prad=float(input("Planetary radius(Earth radii) : "))
koi_prad_err1=float(input("Planetary radius upper unc. :"))
koi_prad_err2=float(input("Planetary radius lower unc. : "))
koi_teq=float(input("Equillibrium Temperature(k) : "))
koi_insol=float(input("Insolation Flux[Earth flux] : "))
koi_insol_err1=float(input("Insolation Flux upper unc. :"))
koi_insol_err2=float(input("Insolation flux lower unc. : "))
koi_model_snr=float(input("Transit signal to noise : "))
koi_tce_plnt_num=float(input("TCE planet number : "))
koi_tce_delivname=input("TCE delivery : ")
koi_steff=float(input("stellar effective temperature(k) : "))
koi_steff_err1=float(input("Stellar effective temperature upper unc.(k) : "))
koi_steff_err2=float(input("Stellar effective temperature lower unc.(k) : "))
koi_slogg=float(input("Stellar surface gravity(log10(cm/s**2)) : "))
koi_slogg_err1=float(input("Stellar surface gravity upper unc. (log10(cm/s**2)) : "))
koi_slogg_err2=float(input("Stellar surface gravity lower unc. (log10(cm/s**2)) : "))
koi_srad=float(input("stellar radius(solar radii) : "))
koi_srad_err1=float(input("stellar radius upper unc.(solar radii) : "))
koi_srad_err2=float(input("stellar radius lower unc.(solar radii) : "))
koi_ra=float(input("RA(Decimal Degrees) : "))
koi_dec=float(input("DEC(Decimal Degrees) : "))
koi_kepmag=float(input("kepler-band(mag) : "))
koi_input_data= {
    #'koi_score':koi_score,
    #'koi_fpflag_nt': koi_fpflag_nt,
    #'koi_fpflag_ss': koi_fpflag_ss,
    #'koi_fpflag_co': koi_fpflag_co,
    #'koi_fpflag_ec': koi_fpflag_ec,
    'koi_period': koi_period,
    'koi_period_err1': koi_period_err1,
    'koi_period_err2': koi_period_err2,
    'koi_time0bk': koi_time0bk,
    'koi_time0bk_err1': koi_time0bk_err1,
    'koi_time0bk_err2': koi_time0bk_err2,
    'koi_impact': koi_impact,
    'koi_impact_err1': koi_impact_err1,
    'koi_impact_err2': koi_impact_err2,
    'koi_duration': koi_duration,
    'koi_duration_err1': koi_duration_err1,
    'koi_duration_err2': koi_duration_err2,
    'koi_depth': koi_depth,
    'koi_depth_err1': koi_depth_err1,
    'koi_depth_err2': koi_depth_err2,
    'koi_prad': koi_prad,
    'koi_prad_err1': koi_prad_err1,
    'koi_prad_err2': koi_prad_err2,
    'koi_teq': koi_teq,
    'koi_insol': koi_insol,
    'koi_insol_err1': koi_insol_err1,
    'koi_insol_err2': koi_insol_err2,
    'koi_model_snr': koi_model_snr,
    'koi_tce_plnt_num': koi_tce_plnt_num,
    'koi_tce_delivname': koi_tce_delivname,
    'koi_steff': koi_steff,
    'koi_steff_err1': koi_steff_err1,
    'koi_steff_err2': koi_steff_err2,
    'koi_slogg': koi_slogg,
    'koi_slogg_err1': koi_slogg_err1,
    'koi_slogg_err2': koi_slogg_err2,
    'koi_srad': koi_srad,
    'koi_srad_err1': koi_srad_err1,
    'koi_srad_err2': koi_srad_err2,
    'ra': koi_ra,
    'dec': koi_dec,
    'koi_kepmag': koi_kepmag
}
input_df = pd.DataFrame([koi_input_data])
input_df_encoded = hii.transform(input_df)
input_df_encoded = pd.DataFrame(input_df_encoded, columns=hii.get_feature_names_out(), index=input_df.index)
input_df_encoded[numeric_cols] = imputer.transform(input_df_encoded[numeric_cols])
prediction=model.predict(input_df_encoded)
print("Prediction : ",prediction)
if(prediction[0]==1):
  print("Confirmed")
elif(prediction[0]==2):
  print("Candidate")
else:
  print("False Positive")
```
